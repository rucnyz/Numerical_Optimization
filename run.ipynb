```python
from sympy import init_printing
from utils import algo_display, get_fun
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
```

```python
plt.rcParams["font.sans-serif"] = ["SimHei"]  # 用来正常显示中文标签
plt.rcParams["axes.unicode_minus"] = False  # 用来正常显示负号
```

```python
# 设置x的维度
n = 5
```

## Watson函数的各算法结果

```python
x0 = np.zeros((n, 1))
fun, grad, Hessan = get_fun("Watson", 5)
print("Watson函数使用各算法对比：\n")
delta_d, delta_m, delta_S, delta_D, delta_B = algo_display(fun, grad, Hessan, x0)
```

```python
plt.figure(dpi=120)
plt.plot(delta_m, label="修正牛顿法")
plt.plot(delta_S, label="拟牛顿法-SR1")
plt.plot(delta_D, label="拟牛顿法-DFP")
plt.plot(delta_B, label="拟牛顿法-BFGS")
plt.legend()
plt.ylim(0, 0.1)
```

## DBV函数的各算法结果

```python
x0 = np.array([((i + 1) / (n + 1)) * ((i + 1) / (n + 1) - 1) for i in range(n)])
x0 = x0.reshape((n, 1))
fun, grad, Hessan = get_fun("DBV", 5)
print("DBV函数使用各算法对比：\n")
delta_d, delta_m, delta_S, delta_D, delta_B = algo_display(fun, grad, Hessan, x0)
```

```python
plt.figure(dpi=120)
plt.plot(delta_m, label="修正牛顿法")
plt.plot(delta_S, label="拟牛顿法-SR1")
plt.plot(delta_D, label="拟牛顿法-DFP")
plt.plot(delta_B, label="拟牛顿法-BFGS")
plt.legend()
plt.ylim(0, 0.1)
```
