```python
# 得到原函数、导数、海塞矩阵表达式
def get_fun(fun_name = "Watson", n = 5):
    assert fun_name in ["Watson", "DBV"]
    f, fun_sym, vector_x = eval(fun_name)(n)
    print('function:')
    pprint(fun_sym, use_unicode = True)
    g = get_gradient(fun_sym, vector_x, True)
    G = get_Hessan(fun_sym, vector_x, True)
    return f, g, G
```
